```python
categories = ['normalizededucation', 'normalizedhealth', 'normalizedweather', 'normalizedcrime', 'normalizedrent']
```

```python
scores = {}
def makeScores(rankings):
    totalRanking = 0
    for index, row in finalColumns.iterrows():
        print("ROW")
        print()
        print(row)
        for i, j in zip(rankings, categories):
            totalRanking+=int(i)*row[j]
            print(int(row[j]))
        scores[row['state']] = totalRanking
        totalRanking = 0
makeScores([1,3,2,1,5])
print(scores)
        
print(scores.keys())
```

```python
# import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import shapely.geometry as sgeom

import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader



def main():
    fig = plt.figure()
    # to get the effect of having just the states without a map "background"
    # turn off the background patch and axes frame
    ax = fig.add_axes([0, 0, 1, 1], projection=ccrs.LambertConformal(),
                      frameon=False)
    ax.patch.set_visible(False)

    ax.set_extent([-125, -66.5, 20, 50], ccrs.Geodetic())

    shapename = 'admin_1_states_provinces_lakes'
    states_shp = shpreader.natural_earth(resolution='110m',
                                         category='cultural', name=shapename)
    print(states_shp)


    ax.set_title('US States which intersect the track of '
                 'Hurricane Katrina (2005)')

    reader = shpreader.Reader(states_shp)
    states = reader.records()

    
    
    def colorize_state(stateAb):
        stateScore = scores[stateAb]
        color = 'white'
        if(stateScore < -9):
            color= "#6c1313"
        elif(-7 > stateScore >= -9):
            color= "#b62020"
        elif(-4 > stateScore >= -7):
            color= "#de7621"
        elif(-1 > stateScore >= -4):
            color= "#dece21"
        elif(2 > stateScore >= -1):
            color= "#de7621"
        elif(5 > stateScore >= 2):
            color= "#9cde21"
        elif(7 > stateScore >= 5):
            color= "#21de8f"
        elif(9 > stateScore >= 7):
            color= "#21ded8"
        elif(stateScore > 9):
            color="#217cde"
        return color;


    for state in states:
        postalAb = state.attributes['postal']
        ax.add_geometries([state.geometry],
                         ccrs.PlateCarree(),
                         facecolor = colorize_state(postalAb),
                         edgecolor = 'black')
        

    # make two proxy artists to add to a legend
    direct_hit = mpatches.Rectangle((0, 0), 1, 1, facecolor="red")
    within_2_deg = mpatches.Rectangle((0, 0), 1, 1, facecolor="#FF7E00")
    labels = ['State directly intersects\nwith track',
              'State is within \n2 degrees of track']
    ax.legend([direct_hit, within_2_deg], labels,
              loc='lower left', bbox_to_anchor=(0.025, -0.1), fancybox=True)

    plt.show()


if __name__ == '__main__':
    main()
```
